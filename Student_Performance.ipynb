## Simple Exploratory Data Analysis

#### Lets take a look at Final Grades 'G3' 

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.bar(stud_mat['G3'].value_counts().index,stud_mat['G3'].value_counts().values,fill = 'navy', edgecolor = 'k', width = 1)
plt.xlabel('G3'); plt.ylabel('Count'); plt.title('Distribution of Final Grades');
plt.xticks(list(range(0, 21)));
```

### Findings:
#### Final Grades are pretty much Normally Distributed

### Now it's Time to Find more Insights by Visualizing it.

#### We going to Visualize each column with respect to our Target (Response) i.e. 'G3'.
#### This approach will help us to understand relation between each column & Target.

### School Column

```python
df_sch=stud_mat[['G3','school']].copy()

df_sch1=df_sch.groupby(['school']).count()

df_sch2=df_sch.groupby(['school']).mean()

print ('Number of Students from each School\n', df_sch1.head())
print ('\nAverage Final Grade of Students from each School\n',df_sch2.head(),'\n\n')


#Plot

ax1 = df_sch1.plot(kind='bar',figsize=(8,5), legend=True, fontsize=12)
ax1.set_title("School (GP/MS) vs Students", fontsize=15)
ax1.set_xlabel("GP & MS", fontsize=12)
ax1.set_ylabel("No. Students", fontsize=12)


ax2 = df_sch2.plot(kind='bar',figsize=(8,5), legend=True, fontsize=12)
ax2.set_title("School (GP/MS) vs Average G3", fontsize=15)
ax2.set_xlabel("School", fontsize=12)
ax2.set_ylabel("Avg. G3", fontsize=12)

plt.show()
```

### Findings:
#### Number of Students from 'GP' (Gabriel Pereira) School are more than 'MS' (Mousinho da Silveira) School.
#### Average of Final Grades of Students from 'GP' (Gabriel Pereira) School is more than 'MS' (Mousinho da Silveira) School.

## Lets Begin Model Building
### We will use Multiple Linear Regression
### Approach (Cascaded Multiple Linear Regression):
#### - Build First Multiple Linear Regression Model with target 'G1' & 'G2'
#### - Build Second Multiple Linear Regression Model by predictors 'G1' & 'G2' and target value 'G3'.

### Creating DataFrame to Predict 'G1'

```python
dataset1 = stud_mat[['age', 'Medu','Fedu', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime',
                     'goout', 'Dalc', 'Walc', 'health', 'absences', 'higher', 'G1']].copy()

dataset1.replace('yes', 1, inplace=True)
dataset1.replace('no', 0, inplace=True)

dataset1.head()
```

```python
# Generating Predictors and Response arrays

X1 = dataset1.iloc[:, :-1].values  
y1 = dataset1.iloc[:, 1].values  
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression  

# Splitting Data into Train & Test sets

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=0)  
```

```python
# Fitting Linear Regression Model

regressor1 = LinearRegression()  

regressor1.fit(X_train1, y_train1)
```

```python
#Predicting values for Test Dataset

y_pred1 = regressor1.predict(X_test1)  

df1 = pd.DataFrame({'Actual': y_test1, 'Predicted': y_pred1})

df1.head()
```

### Metrics for Evaluating Model Performance

```python
from sklearn import metrics

mae = metrics.mean_absolute_error(y_test1, y_pred1)
print('Mean Absolute Error:', mae)  

mse = metrics.mean_squared_error(y_test1, y_pred1)
print('Mean Squared Error:', mse)

rmse = np.sqrt(metrics.mean_squared_error(y_test1, y_pred1))
print('Root Mean Squared Error:', rmse)
```

### First Model Performance to Predict 'G1':
#### Mean Absolute Error: 9.42
#### Mean Squared Error: 3.27
#### Root Mean Squared Error: 1.80

### Creating Dataset & Building Model to Predict 'G2'

```python
dataset2 = stud_mat[['age', 'Medu','Fedu', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime',
                     'goout', 'Dalc', 'Walc', 'health', 'absences', 'higher', 'G2']].copy()

dataset2.replace('yes', 1, inplace=True)
dataset2.replace('no', 0, inplace=True)


X2 = dataset2.iloc[:, :-1].values  
y2 = dataset2.iloc[:, 1].values  

# Splitting Data
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=0)  

#Fitting Model
regressor2 = LinearRegression()  
regressor2.fit(X_train2, y_train2)

#Predicting
y_pred2 = regressor2.predict(X_test2)  

df2 = pd.DataFrame({'Actual': y_test2, 'Predicted': y_pred2})

df2.head()
```

### Metrics for Evaluating Performance of Model

```python

mae = metrics.mean_absolute_error(y_test2, y_pred2)
print('Mean Absolute Error:', mae)  

mse = metrics.mean_squared_error(y_test2, y_pred2)
print('Mean Squared Error:', mse)

rmse = np.sqrt(metrics.mean_squared_error(y_test2, y_pred2))
print('Root Mean Squared Error:', rmse)
```

### First Model Performance to Predict 'G2':
#### Mean Absolute Error: 9.42
#### Mean Squared Error: 3.27
#### Root Mean Squared Error: 1.80
